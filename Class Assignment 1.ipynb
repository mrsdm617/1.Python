```python
class OddEven():
    def oddeven():
        num=int(input("Enter the number:"))
        if((num%2)==0):
            print(num," is Even Number")
            message=num," is Even Number"
        else:
            print(num," is Odd Number")
            message=num," is Odd Number"
            return message
```
